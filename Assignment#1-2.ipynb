## 주문서를 보고 번호를 입력하면, 해당 작업을 수행하는 무인 카페 프로그램을 작성

```python
class Menu():
    def __init__(self):
        self.orderList = {}
        self.total = 0
        self.f = open('cafe.txt', 'r', encoding='UTF8')
        self.__addMenu(self.f)
        
    
    def __addMenu(self, f):
        # file을 받아서 한줄씩 읽어서 List에 추가
        while(True):
            line = f.readline()
            if not line:
                break
            word = line.split(',')
            word[3] = word[3].replace("\n", "")
            self.orderList[word[0]] = [word[1],word[2],word[3]]
            
    
    def printMenu(self):
        # 전체 메뉴를 출력
        print("===================")
        for key, items in self.orderList.items():
            print("%s. %s  금액: %s원 재고: %s"%(key, items[0],items[1],items[2]))
        print("end를 입력하시면 주문서로 돌아갑니다. ")
        print("===================")
```

## Exercise1. Create a class "Queue" that stores integers based on the list. There are 5 methods, such as enqueue, dequeue, peek, isEmpty, and queueSize in a Queue class.

```python
class Queue():
    def __init__(self):
        self.queue = []
        
    def enqueue(self, x):
        self.queue.append(x)
    
    def dequeue(self):
        num = None
        
        if not self.isEmpty():
            num = self.peek()
            del self.queue[0]
            
        return num
    
    def peek(self):
        peek_data = None
        
        if not self.isEmpty():
            peek_data = self.queue[0]
            
        return peek_data
    
    def isEmpty(self):
        if len(self.queue)==0:
            return True
        else: 
            return False
    
    def queueSize(self):
        return len(self.queue)
```

```python
from random import *

i = 0
Q = Queue()

while i<10:
    num = randint(1, 2)
    
    print("---%d step ---"%(i+1))
    print("operation:", num)
    if num==1:
        Q.enqueue(i)
    else:
        num = Q.dequeue()
    
    print("Current queue: ", Q.queue)
    print("Peek data: ", Q.peek())
    print("Queue size: ", Q.queueSize())
    
    i+=1
```

## Exercise2. There is a 4-tier shelf 4 spaces on each ties. We call this a Shelf class. The description of each method in a shelf class are as follows.

```python
class Shelf():
    def __init__(self):
        self.shelf = [[],[],[],[]]
    
    def addNum(self, x):
        if x%4==0:
            if not self.isFull(1):
                self.shelf[0].append(x)
            else:
                print("1층은 꽉차서 더이상 공간이 없습니다.")
    
        elif x%4==1:
            if not self.isFull(2):
                self.shelf[1].append(x)
            else:
                print("2층은 꽉차서 더이상 공간이 없습니다.")
        elif x%4==2:
            if not self.isFull(3):
                self.shelf[2].append(x)
            else:
                print("3층은 꽉차서 더이상 공간이 없습니다.")

        elif x%4==3:
            if not self.isFull(4):
                self.shelf[3].append(x)        
            else:
                print("4층은 꽉차서 더이상 공간이 없습니다.")

    
    def delNum(self, index):
        if index==-1:
            if not self.isEmpty(1):
                del self.shelf[0][len(self.shelf[0])-1]
            else:
                print("1층은 비어서 제거할 숫자가 없습니다.")
        if index==-2:
            if not self.isEmpty(2):
                del self.shelf[1][len(self.shelf[1])-1]
            else:
                print("2층은 비어서 제거할 숫자가 없습니다.")
        if index==-3:
            if not self.isEmpty(3):
                del self.shelf[2][len(self.shelf[2])-1]
            else:
                print("3층은 비어서 제거할 숫자가 없습니다.")
        if index==-4:
            if not self.isEmpty(4):
                del self.shelf[3][len(self.shelf[3])-1]
            else:
                print("4층은 비어서 제거할 숫자가 없습니다.")
    
    
    def levelSize(self, level):
        return len(self.shelf[level-1])
    
    def isFull(self, level):
        if self.levelSize(level)==4:
            return True
        return False
    
    def isEmpty(self, level):
        if self.levelSize(level)==0:
            return True
        return False
```

```python
S = Shelf()
num = 0
print("-4 <= N <= -1: 각 층의 최근 입력숫자 제거, 나머지 음수: 종료")
while num > -5:
    num = int(input("0혹은 양수 입력: "))
        
    if num >= -4 and num <=-1:
        S.delNum(num)
        
    elif num < -4:
        print("종료합니다.")
        break
        
    else:
        S.addNum(num)
    
    print("--> 현재박스상태", S.shelf)
```

## Exercise3. create a class to find the area of a square and the common area of two squares

```python
import math
a=0
class Square:
    def __init__(self, cx, cy, cr):
        self.cx = cx
        self.cy = cy
        self.cr = cr
        
    def center(self):
        return (self.cx, self.cy)
    
    def width(self):
        w = self.cr ** 2
        return w
        
class Common_area(Square):
    
    def compute_common_area(self, other):
        x1, y1 = self.cx-self.cr/2, self.cy+self.cr/2 
        x2, y2 = self.cx+self.cr/2, self.cy-self.cr/2 
        x3, y3 = other.cx-other.cr/2, other.cy+other.cr/2
        x4, y4 = other.cx+other.cr/2, other.cy-other.cr/2

        ## case1 other squarej0 l$k%8l*=l
        if x2 < x3:
            return 0

        ## case2 other squarej0 l
        if x1 > x4:
            return 0

        ## case3 other squarej0 lkl*=l
        if  y2 > y3:
            return 0

        ## case4 other squarej0 ll*=l
        if  y1 < y4:
            return 0

        left_up_x = max(x1, x3)
        left_up_y = min(y1, y3)
        right_down_x = min(x2, x4)
        right_down_y = max(y2, y4)

        width = right_down_x - left_up_x
        height =  right_down_y - left_up_y

        return abs(width * height)
      
```
